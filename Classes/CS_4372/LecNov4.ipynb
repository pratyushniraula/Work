```python
#language model: predict next word based on previous words
#given a prompt the model generates text one word at a time
#sometimes language models are autoregressive

#watch 3blue1brown transformers

#n-gram models
#neural network based language models that use fixed size context windows
#given previous n words, predict the next word
#sliding window approach
#limitations: fixed context window size, data sparsity
#addressing limitations with embeddings and rnn/transformers
#example of n-gram model: predicting next word after "the cat sat on the"
#unigram model: looks at last word "the" -> predicts "mat" (most frequent following word)
#bigram model: looks at last two words "on the" -> predicts "mat"
#higher order n-grams capture more context but require more data
#smoothing techniques (like add-one smoothing) help with data sparsity
#evaluate n-gram models using perplexity
```

```python
#recurrent neural network
#process sequences of arbitrary length
#shared weights across time steps (hidden layers have same weights)
#maintain hidden state that captures information about previous words
#model size does not grow with sequence length, but it is a little slow since you have to do learning at every step

#search up rnn style language models
#given a sequence of T words, you can predict the T+1 word

#can theoretically capture long-range dependencies
#train using backpropagation through time (BPTT)
#limitations: vanishing/exploding gradients, training complexity
#addressing limitations with LSTM/GRU units and transformers




#unlike a vanilla nn model, you take one word into one hidden layer at a time, update hidden state, then take next word and so on

#rnns can have an output at every step or just the last step
#if you have outputs at every step, you can see the error and loss at every step and backpropagate through time accordingly
```

```python
#in every word in english it could be noun verb adjective etc.

#a pos tagger labels each word with its part of speech
#rnn can be used for pos tagging by having an output at every step

#one output at the last step example: sentiment analysis
#given a movie review, predict if its positive or negative sentiment
#process the review word by word, then use the final hidden state to predict sentiment

#modern example: give a whole article, predict if its pro or against something

#example of having output at every step: predicting the next word in a sentence
#given a sequence of words, predict the next word at each time step


#teacher forcing: during training, use the actual previous word as input for the next time step instead of the predicted word
#helps the model learn faster and more accurately
#during inference, use the model's own predictions as input for the next time step


#you can compute error at the last step only or at every step
#computing error at every step provides more feedback and can lead to better learning
#but it can also be more computationally expensive

#for rnn backpropogation you use multivariable chain rule to compute gradients through time
#you sum the gradients from all time steps to update weights
#bptt: backpropagation through time
# used to train a rnn through time: time refers to the steps not actual time


#example: train a nn on shakespear or obama and have it generate text in that style
#start with a prompt and have the model generate text one word at a time
#can use temperature parameter to control randomness of generated text
#low temperature -> more conservative, high probability words
#high temperature -> more random, diverse words
#sampling from the output distribution to get next word

#perplexity: measure of how well a probability model predicts a sample
#lower perplexity indicates better predictive performance (less confused) (how surprised the model is)
#perplexity = 2^cross-entropy
#perplexity basically is an evaluation metric, that measures how well your language model is predicting the next word in a sequence
#probability is based on the probability a model assigns to a sequence of words



#ngram approach is oldschool approach, and rnns are much better and more modern
#but rnns have a problem, vanishing gradient problem: backpropogation through many time steps causes gradients to vanish or explode
    #signal gets weaker and weaker as you go back in time steps (travelling from the end to the beginning)
#or you could have exploding gradients where the signal gets too strong, example: weights get too large and cause instability at training for deep networks

#vanilla rnns have vanishing gradient, addressed with lstm and gru units

#lstm: long short term memory
#special rnn unit designed to capture long-range dependencies
#apple introduced lstm in the wwdc 2016 keynote

#lstm has cell state and hidden state
#cell state carries long-term information
#hidden state carries short-term information
#gates control the flow of information into and out of the cell state
#forget gate: decides what information to discard from cell state
#input gate: decides what new information to add to cell state
#output gate: decides what information to output from cell state to hidden state   

#think of it as long term being carried in a conveyor belt, and short term being the stuff you are working on right now

#cell state is the sum of all the important information you have seen so far
#gates are like filters that decide what information to keep or discard

#the first gate is the forget gate, which decides what information to discard from the cell state
#it looks at the previous hidden state and the current input, and outputs a value between
#0 and 1 for each number in the cell state
#it tries to forget unimportant information to make room for new information
#the gate will take from the the input gate and the previous hidden state to decide what to forget in the cell state


#then you have the addition gate which adds something new to the cell state


#and so you forget something, you add something and then you output something from the cell state to the hidden state

#long term memory is at the top from the past that you are constantly updating and revising to make sure it is current. 
#The bottom is the short term memory that you are working on right now
#lstm can capture long-range dependencies better than vanilla rnn
#but it is more complex and computationally expensive

#you can remember things from a long time ago better with lstm than vanilla rnn by controlling the flow of information with gates

#in vanilla rnn you can only get 7 time steps back, while in lstm you can get 100 time steps back

#bidirectional lstm: processes the sequence in both forward and backward directions
#captures context from both past and future words
#useful for tasks like pos tagging and named entity recognition
#for example: the movie was terribly exciting!
#the word terribly can have different meanings based on context
#bidirectional lstm can capture context from both sides to disambiguate meaning
```

```python
#lstm still had room for improvement, sequence to sequence model improved things like machine translation further

#seq2seq model: consists of an encoder rnn and a decoder rnn
#rather than remembering everything one at a time, 
#encoder encodes the encoding of the entire source sentence, which is passed into the input to the decoder rnn
#where it generates target entence conditioned on the source sentence encoding
#this became popular to translation, summarization, dialogue, code generation, and more

#after lstm seq2seq became popular
#then they made multilayer seq2seq
```

```python
#the revolution that came was the attention mechanism 
#a special mechanism called the attention mechanism that allows the decoder to look back at the entire source sentence encoding at every time step
#cross attention: the start step of the decoder attends to all the encorder hidden states 
#   (pays attention to all the hidden states of the encoder and decides which are important)
#the first step of the decoder says that the first step of the encoder matters to me
#then the second step of the decoder says that the nth step of the encoder matters to me
#and so on
```
