```python
#most primitive approach to language models is n-gram

#n-gram 
    #1: unigram (bag of words)
    #2: bigram
    #3: trigram

#2: logistic regression/naive bayes

#3: RNN
    #can process any length input sequence
    #computation for step t can use information from many steps back
    #there is a hidden state passed from one step to the next to the next
    #cannot learn long term dependencies

#4: LSTM
    #look more into it from colahs blog for lstm
    #have cell states that you can add or forget information in order to remember long range information
    #can remember up to 100 steps back
    #bidirectional lstm: process input sequence in both directions
    #gru (gated recurrent unit): similar to lstm but simpler, combines forget and input gates into a single update gate, has fewer parameters, faster to train, performs similarly to lstm in many tasks

    #what problem does lstm have? sequentially passing information forward

#5: seq-2-seq
    #has two models: encoder and decoder
    #encoder encodes the input sequence into a fixed length vector
    #decoder decodes the fixed length vector into an output sequence
    #problem: fixed length vector is a bottleneck, hard to capture long sentences or complex structures

#6: attention mechanism
    #decoder encoder attention: decoder attends to all encoder hidden states at every time step
    #decoder will not just depend on the last step of the encoder, it will pay attention to every step in the encoder
        #take dot product and apply softmax to get weights so that you can get a weighted sum of the encoder hidden states and see which parts matter most
            #whatever matters most will have the highest weight
    #after seeing what matters the most from the encoder for one of the decoder steps, it will figure out what step of the encoder matters most

#then mfs started focusing more on attention mechanism
#attention provides a more human like way of processing language

#other attention mechanisms:
```
