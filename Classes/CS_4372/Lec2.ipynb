```python
# craps example

import numpy as np
import numpy.linalg

dice1 = np.random.randint(1, 7, 10) # 10 random integers from 1 to 6
dice2 = np.random.randint(1, 7, 10)
print(dice1)
print(dice2)

sum = dice1 + dice2
print(sum)
#min sum is 2, max sum is 12. These are rare numbers since they both require 1 and 1 or 6 and 6
#what about finding the most common one?
```

```python
dice1 = np.random.randint(1, 7, 10000) # 10 random integers from 1 to 6
dice2 = np.random.randint(1, 7, 10000)
print(dice1)
print(dice2)
sum = dice1 + dice2
print(sum)
```

```python
#histogram plot: y axis is frequency, x axis is the value
from matplotlib import pyplot as plt
```

```python
plt.hist(sum)

#seems as though the number 7 is the most common but we can improve the visualization
```

```python
plt.xticks(range(2, 13)) #set x axis ticks to be from 2 to 12
plt.hist(sum, bins=11, range=(2, 13), edgecolor="black", density=True) #11 bins from 2 to 13, density true means y axis is percentage
```

```python
#create a 2d array of size (3, 4) containing numbers 1-12
array_2d = np.arange(1, 13).reshape(3, 4) #arrange creates a 1d array from 1 to 12 (13-1 is 12), reshape changes it to 3 rows and 4 columns
print(array_2d)
```

```python
array2_2d = np.arange(1, 13).reshape(3, 4, order='F') #arrange creates a 1d array from 1 to 12 (13-1 is 12), reshape changes it to 3 rows and 4 columns
print(array2_2d)

#default order is C which is row major order, F is column major order
```

```python
x = np.arange(1,5).reshape(2,2)
print(x, end="\n\n----\n\n")
y = np.arange(5,9).reshape(2,2)
print(y)
```

```python
#matrix multiplication

np.dot(x, y) #3x4 dot 2x2 is not possible
print(np.dot(x,y))
print("", end="\n-----\n")
#or you can do x @ y
x @ y
print(x@y)
print("", end="\n-----\n")
x*y #element wise multiplication
```
