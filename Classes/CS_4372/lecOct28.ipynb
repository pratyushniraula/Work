```python
#we are working with tensorflow  which is a good library for deep learning
#this library is in competition with pytorch

#the reason we work with tensorflow is also because we have access to the keras api and we have access to the tfhub which has pre trained models

#in colab when importing the tensorflow library it automatically uses the gpu if available
#whenever you use tf.keras.util.get_file it caches the file in the local drive so that it does not have to be downloaded again
#this is found in root/.keras/datasets

#tensorflow has a dataset format

#get familiar with the tf.keras.utils library

#it has something called image_dataset_from_directory which is very useful for loading images from a directory structure

#the number being trained when doing epochs is referred to batch size


#tensorflow has data augmentation and dropout support built in, we need to use it to prevent overfitting and on our assignment
#we use data augmentation to increase the size of our dataset artificially by transforming the images slightly

#just as for any hyperparameter we decide how to tune the dropout rate and data augmentation parameters

#dropout does not reduce the number of parameters but it reduces overfitting by randomly dropping connections during training
#dropout is generally for dense layers
#suppose it is a dense connection 5x5 = 25 connections
#with dropout rate of 0.2 we randomly drop 20% of the connections during training
#during testing we use all the connections but we scale down the weights by 0.8 to account for the missing connections during training
#we temporarily set 5 connections to 0 during training
#in every batch in training we select a random set of connections to drop
#in the next batch that we might drop a different set of connections
#all connections are still there but we are just not using them during training

#to reiterate: dropout does not permanantly reduce the number of parameter, it temporarily drops connections during training to reduce overfitting
```

```python
#to find the number of parameters in a later we do
#(number of input neurons + 1) * number of output neurons
#input neurons: number of features
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
print(32*3*3*1 + 32)  # First Conv2D layer parameters
print((3*3) * 64)   # Second Conv2D layer parameters
print((64 * 7 * 7 + 1) * 64)  #Dense layer parameters
print((64 + 1) * 10)   # Output layer parameters
```
