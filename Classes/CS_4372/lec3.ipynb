#high school notation
y = a + bx

y = house price
x = house size

ml notation
y = w0 + w1 * x



take a dataset and find the w's

the set of w's {w} -> parameters, weights


given a dataset you wnt to find the best w's

your model wants to match the actual y as much as possible
    x                   y
    (size)              (actual)
1   1000                600,000
2   1200                700,000


we must define an error function:

mean squared error is an example
mse = 1/N * (sum(h_i - y_i)^2)       sum: i = 1, up to N
    we define the function as h(i) = w0 + w1 * (x^(i))                  // i is the index not a power
    y^(i) is the actual
    unknowns are w0 and w1
    mse = 1/N * (sum(w0 + w1 * x^(i) - y^(i))^2)
    find {w} that minimizes MSE
        partial derivative(J/w0) = 0
        partial derivative(J/w1) = 0
        
        #these are the functions we would get from minimizing the derivatives.
        a1w0 + b1w1 = c1
        a2w0 + b2w1 = c2
        and so we solve for the solutions to solve for w
        |a1 b1| |w0 | = |c1|
        |a2 b2| |w1 | = |c2|
        this is a matrix where A*W = C

        you want to solve for w matrix by taking the inverse so that you move something to the other size to make it W = A^-1 * C

        time complexity of matrix inversion using gaussian elimination is O(n^3)
        

        in deep learning there could be a chance to have hundreds of thousands of parameters
        is the alternative to doing the partial derivatives part? we found something called gradient descent
            machine learning can be described as gradient descent applied onto different things
            gradient descent gives you a linear way to solve things (O(n)), but it does not garuntee the best solutions
                it could be a local best solution
            gradient descent also has to assume convexity and not garunteed convergence
            problem is function J must be convex, smooth, and differentiable
            and there is something called the learning rate (step size) where you may or may not get a convergence for the step size that's not tuned (too large or too small)

            but what we can do is instead of inversing a matrix (O(n^3)) we can use gradient descent so that we can do it in O(n)

```python
#pandas library
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
frame = pd.DataFrame(data)
print(frame.head()) #prints the first 5 rows of the dataframe
print(frame.tail()) #prints the last 5 rows of the dataframe
print(frame.shape) #prints the shape of the dataframe (number of rows and columns)
print(frame)
```

```python
#with a dataframe learn how to do selection, projection, group by and aggregation
#selection: selecting rows based on some condition
print(frame[frame['species'] == 'setosa']) #selecting all rows where species is setosa
# or
frame[frame['sepal_length'] > 5.0] #selecting all rows where sepal_length is greater than 5.0
# if you do just frame['sepal_length'] > 5.0 it will return a boolean array

#projection: selecting specific columns
print(frame[['sepal_length', 'sepal_width']]) #selecting only sepal_length and sepal_width columns
#or
frame['sepal_length'][frame['sepal_length'] > 5.0] #selecting sepal_length column where sepal_length is greater than 5.0
#or
frame[frame['sepal_length'] > 5.0 & (frame['species'] == 'setosa')] #selecting sepal_length column where sepal_length is greater than 5.0 and species is setosa
#or 
frame.loc[0:2, ['sepal_length', 'species']] #selecting sepal_length and species columns for rows 0 to 2
#or
frame.iloc[0:2, [0, 4]] #selecting sepal_length and species columns for rows 0 to 2 using index
#if you type loc it will include the last index but if you type iloc it will not include the last index



#group by and aggregation: grouping by species and calculating the mean
print(frame.groupby('species').mean()) #grouping by species and calculating the mean of all other columns
print(frame.groupby('species')['sepal_length'].mean()) #grouping by species and calculating the mean of sepal_length column only

#or
#columns can be of 3 types: numerical, categorical (can you group by), text(string [not importang to us])

frame.groupby('species').mean()
frame.groupby('species')['sepal_length'].std() #standard deviation

#or
frame.groupby('species')['sepal_length'].agg([np.mean, np.std, np.min, np.max]) #aggregating multiple functions at once








frame['sepal_length'].value_counts() #gives you the count of each unique value in the sepal_length column
frame.describe() #gives you the summary statistics of the dataframe: count, mean, std, min, 25%, 50%, 75%, max for each numerical column

#find the mean of sepal_length for each species where sepal_width is greater than 3.0: this requires selection, projection, group by and aggregation
frame[frame['sepal_width'] > 3.0].groupby('species')['sepal_length'].mean()
```
