xgboost:
    stands for extreme gradient boosting
    xgboost has a history. it started in the university of washington from a phd student. it was so popular that he made it open source
    microsoft made their own model called lightGBM: light gradient boosting machine

    the main purpose of this, it makes an objective function that deals with overfitting: minimizing training loss (training error (how well model fit on training data)) and regularization (measures complexity of model)
        Obj(theta) = L(theta) + omega(theta)
        optimizing training loss encourages predictive models... (find in slides)

    regression tree (also known as classification and regression tree) he puts everything in a tree where root and internal nodes are decisions and leafs have scores in them
    regression tree ensemble combines multiple trees and take a net score (adding up those scores)
