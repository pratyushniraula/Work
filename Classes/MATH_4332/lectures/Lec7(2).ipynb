```python
# String methods:
# Those are functions that return information about the string or
# return a new string that is a modified version of the original.
# Methods are found by using dot after string's name.
# E.g
s = "Hello! There it is."
s.lower() # returns a string with every letter of the original in lower characters.
print(s.lower())
s.upper() # returns a string with every letter of the original in upper characters.
print(s.upper())
s.replace("e","a") # returns a string with every occurrence of e replaced by a.
print(s.replace("e","a"))

#print(s)
s.count("e") # This counts the number of occurrences of e
print(s.count("e"))
s.index("e") # returns the location of the first occurence of e
print(s.index("e"))
# The isalpha method is used to tell if a character is a letter or not.
s.isalpha() # returns true if every character of s is a letter
print(s.isalpha())
print(s.replace("e","a",1)) # replace first e by a
```

```python
# The lower,upper and replace methods do not change
# the original string. They rather create a new string.
# How to replace the original string?
s = s.lower() # a new string s.lower() is created and assigned to s
s = s.upper() # a new string s.upper() is created and assigned to s
s = s.replace() # a new string s.replace() is created and assigned to s
```

```python
# isalpha can be used in an if conditions.
s = input("Enter a string: ")
if s[0].isalpha():
  print("String starts with letter")
else:
  print("String starts with a non-letter")
```

```python
# We can also check if string contains non letter using not operator
s = input("Enter a string: ")
if not s.isalpha():
  print("String contains non-letters")
```

```python
# If you try to find the index of something that is not in a string,
# Python will raise an error.
# For instance, if
s = "abcde"
# and you try
#s.index("x") # you will get an error
# One way around this is to check first, like below:
if "x" in s:
  location = s.index("x")
  print(location)

if "d" in s:
  location = s.index("d")
  print(location)
```

```python
#6
s = input("Enter a text: ")
# Convert the string to lowercase
s = s.lower()
# Remove spaces
for char in " ":
  s=s.replace(char,"")
# Remove non-alphanumeric characters
for char in s:
  if not char.isalnum():
    s=s.replace(char,"")

if s==s[::-1]:
  print(s,"is a palindrome")
else:
  print(s,"is not a palindrome")
# Enter:
# A man, a plan, a canal, Panama.
# Euston saw I was not Sue.
# Never a foot too far, even.
# Live on evasions? No, I save no evil.
# Gate-man sees name, garage-man sees name-tag.
# Live not on evil.
```
