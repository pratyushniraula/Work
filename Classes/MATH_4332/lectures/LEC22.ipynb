```python
# Regular expressions:
# Raw strings
# Python raw string treats backslash (\) as a literal character,
# which is particularly beneficial when dealing with strings
# containing many backslashes, such as Windows paths or
# regular expression patterns.
# To mark a string as a raw string,
# preface it with an r(or an R) like below:
s = 'a\tb\nA\tB'
print(s)
# a b
# A B
s1 = "line1\nline2"
print(s1)
# line1
# line2
s2 = r"line1\nline2"
# s2 = "line1\\nline2"
print(s2)
# line1\nline2
s3 = r'a\tb\nA\tB'
# s3 = "a\\tb\\nA\\tB"
print(s3)
# 'a\tb\nA\tB'
s4 = r'This is a raw string. Backslashes do not do anything special.'
print(s4)
# This is a raw string. Backslashes do not do anything special.
print(s4=='This is a raw string. Backslashes do not do anything special.')
# True
```

```python
# In a regular string, an escape sequence is
# interpreted as a single character.
# However, in a raw string, each backslash is
# considered a separate character.
# Get the length of a string (number of characters) in Python
s = 'a\tb\nA\tB'
print(len(s))
# 7
rs = r'a\tb\nA\tB'
print(len(rs))
print(rs)
# 10
```

```python
# Windows paths
# Raw strings are particularly useful for representing
# Windows paths, which are delimited by backslashes \.
# Instead of having to escape each backslash
# as \\ in a regular string,
# you can use raw strings to write them as is below.
path = 'C:\\Windows\\system32\\cmd.exe'
rpath = r'C:\Windows\system32\cmd.exe'
print(path == rpath)
# True
```

```python
# However, be aware that a string ending with
# an odd number of backslashes will cause an error.
# To avoid this issue, you can either represent
# the entire path as a regular string
# (replacing each backslash with \\) or represent
# the path as a raw string up to the second to last backslash,
# then represent the last backslash
# as a regular string and concatenate them.
path2 = 'C:\\Windows\\system32\\'
rpath2 = r'C:\Windows\system32\'
# SyntaxError: unterminated string literal
rpath2 = r'C:\Windows\system32' +'\\'
print(rpath2==path2)
```

```python
# Repetition Quantifiers
# Code            Description
# +               match 1 or more occurrences (1+)
# *               match 0 or more occurrences (0+)
# ?               match 0 or 1 occurrence     (optional)
# {m}             match exactly m occurrences
# {m,n}           match between m and n occurrences (both inclusive)
# {m,}:           match m or more occurrences (m+)
```

```python
'''
Greediness of Repetition Quantifiers *, +, ?, {m,n}:
The repetition operators are greedy operators,
and by default grasp as many characters as possible for a match.
For example, the regex xy{2,4} try to match
for "xyyyy", then "xyyy", and then "xyy".
'''
```

```python
# Here is an example that matches an A followed by zero or more B’s:
re.sub(r'AB*', '*', 'ABC ABBBBBBC AC')
```

```python
# Here is an example that matches an A followed by zero or one B:
re.sub(r'AB?', '*', 'ABC ABBBBBBC AC')
```

```python
# Here is an example that matches an A followed by exactly three B’s:
re.sub(r'AB{3}', '*', 'ABC ABBBBBBC AC')
```

```python
# Here is an example that matches an A followed by one to three B’s:
re.sub(r'AB{1,3}', '*', 'ABC ABBBBBBC AC')
```

```python
# Here is an example that matches an A followed by at most two B’s:
re.sub(r'AB{,2}', '*', 'ABC ABBBBBBC AC')
```

```python
# Here is an example that matches an A followed by at least one B:
re.sub(r'AB{1,}', '*', 'ABC ABBBBBBC AC')
```

```python
# Here is an example that matches an A followed by three to six B’s:
re.sub(r'AB{3,6}', '*', 'ABB ABBB ABBBB ABBBBBBBBB')
```

```python
# In the last piece, there is an A followed by nine B’s.
# What gets matched is the A along with the first six B’s.
# Note that the matching in the last piece above is greedy;
# that is, it takes as many B’s as it is allowed.
# It is allowed to take between three and six B’s, and it takes all six.
# The greedy match will try to match as many repetitions
# of the quantified pattern as possible to have the
# biggest match result possible.
```

```python
# To get the opposite behavior (non-greedy match), that is
# to get it take as few B’s as allowed, add a ? to the quantifier
# The non-greedy match will try to match as few repetitions
# of the quantified pattern as possible to have the
# smallest match result possible.
```

```python
'''
Lazy Quantifiers *?, +?, ??, {m,n}?, {m,}?, : You can put an extra ?
after the repetition operators to curb its greediness(i.e., stop at the
shortest match).
'''
```

```python
# Examples
re.sub(r'AB+?', '*', 'ABC ABBBBBBC AC')
```

```python
# Preceding and following matches
# Sometimes you want to match things if they are preceded
# or followed by something.
# Code            Description
# (?=)            matches only if followed by
# (?!)            matches only if not followed by
# (?<=)           matches only if preceded by
# (?<!)           matches only if not preceded by
```

```python
import re
# Here is an example that matched the word 'the' only
# if it is followed by ' cat':
re.sub(r'the(?=\scat)', '*', 'the dog and the cat')
```

```python
# Here is an example that matched the word 'the' only
# if it is followed by ' cat':
re.sub(r'the(?= cat)', '*', 'the dog and the cat')
```

```python
# Here is an example that matched the word 'the' only
# if not followed by ' cat':
re.sub(r'the(?!\scat)', '*', 'the dog and the cat')
```

```python
# Here is an example that matched the word 'the' only
# if it is followed by ' dog':
re.sub(r'the(?=\sdog)', '*', 'the dog and the cat')
```

```python
# Here is an example that matches the word 'the' only
# if it is preceded by a space:
re.sub(r'(?<=\s)the', '*', 'Athens is the capital.')
```

```python
# Here is an example that matches the word 'the' only
# if it is preceded by a space:
re.sub(r'(?<= )the', '*', 'Athens is the capital.')
```

```python
# Here is an example that matches the word 'the' only
# if it is preceded by the character A:
re.sub(r'(?<=A)the', '*', 'Athens is the capital.')
```

```python
# The following example will match the word 'the' only
# if it is neither preceded by and nor followed by letters,
# so you can use it to replace occurrences of the word 'the',
# but not occurrences of 'the' within other words.
re.sub(r'(?<![a-zA-Z])[Tt]he(?![a-zA-Z])', '*', 'The cat is on the lathe there.')
```

```python
# Summary
"""
Expression     Description

[]          any of the characters inside the brackets
[^]         any of the characters not inside the brackets
[.-.]       It matches any single character that sorts between the two characters, inclusive.
.           any character except the newline
+           1 or more of the preceding
*           0 or more of the preceding
?           0 or 1 of the preceding
{m}         exactly m of the preceding
{m,n}       between m and n (inclusive) of the preceding
?           following +, *, ?, {m}, and {m,n} — take as few as possible
\           escape special characters
|           “or”
^           (at start of pattern) match just the first occurrence
$           (at end of pattern) match just the last occurrence
\d \D       any digit (non-digit)
\w \W       any letter or number or underscore (non-letter or -number or -_)
\s \S       any whitespace (non-whitespace)
(?=)        only if followed by
(?!)        only if not followed by
(?<=)       only if preceded by
(?<!)       only if not preceded by
(?i)        flag to ignore case
(?s)        flag to make the . matches newlines too
"""
```

```python
# Here are some short examples:
"""
Expression          Description
'abc'               the exact string abc
'[ABC]'             an A, B, or C
'[a-zA-Z][0-9]'     match a letter followed by a digit
'[a..]'             a followed by any two characters (except newlines)
'a+'                one or more a’s
'a*'                any number of a’s, even none
'a?'                zero or one a
'a{2}'              exactly two a’s
'a{2,4}'            two, three, or four a’s
'a+?'               one or more a’s taking as few as possible
'a\.'               a followed by a period
'ab|zy'             an ab or a zy
'^a'                a at start
'a$'                a at the end
'\d'                every digit
'\w'                every letter or number or _
'\s'                every whitespace
'\D'                everything except digits
'\W'                everything except letters, numbers and _
'\S'                everything except whitespace
'a(?=b)'            every a followed by a b
'a(?!b)'            every a not followed by a b
'(?<=b)a'           every a preceded by a b
'(?<!b)a'           every a not preceded by a b
"""
```

```python
# Note:
"""
Most of the special regex characters lose
their meaning inside bracket list,
and can be used as they are; except ^, -, ] or \.
To include a ], place it first in the list, or use escape \].
To include a ^, place it anywhere but first, or use escape \^.
To include a - place it last, or use escape \-.
To include a \, use escape \\.
No escape needed for the other characters such as ., +, *, ?, (, ), {, },
and etc, inside the bracket list.
You can also include metacharacters, such as \w, \W, \d, \D, \s, \S
inside the bracket list.
"""
'''
In regex, the uppercase metacharacter is always the inverse of the lowercase
counterpart.
'''
```

```python
# Greediness, Laziness and Backtracking for Repetition Operators
'''
Greediness:
Greedy quantifiers first tries to repeat the token as many times
as possible, and gradually gives up matches as the engine
backtracks to find an overall match.
'''
'''
Laziness:
Lazy quantifier first repeats the token as few times as required,
and gradually expands the match as the engine backtracks
through the regex to find an overall match.
'''
'''
Backtracking:
If a regex reaches a state where a match cannot be completed,
it backtracks by unwinding one character from the greedy match.
For example,
if the regex z*zzz is matched against the string "zzzz", the z*
first matches "zzzz"; unwinds to match "zzz"; unwinds to match "zz";
and finally unwinds to match "z",
such that the rest of the patterns can find a match.
'''
```

```python
# compile
"""
If you are going to be reusing the same pattern,you can save
a little time by first compiling the pattern, as shown below:
"""
```

```python
string = '= A3+B2+C8'
pattern = re.compile(r'([ABC])(\d)')
print(pattern)
matches = pattern.finditer(string)
for match in matches:
  print(match.span())
  print(match.group())
  print(match.groups())
```
