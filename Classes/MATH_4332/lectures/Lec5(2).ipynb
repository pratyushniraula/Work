```python
# Counting
# Very often we want our program to count how many times
# something happens.
# For instance a math program may want to count how many numbers have
# a special property
```

```python
# Example: Create a program that asks the user to enter 5 numbers
# and count how many of them are greater than 10.
count = 0 # start count at zero
for k in range(5):
  num = float(input("Enter a number: "))
  if num > 10:
    count = count + 1 # Increase the count by 1 every time we get
                      # a number larger than 10
print(count)
```

```python
# We can change count = count + 1 to
# count +=1
count = 0 # start count at zero
for k in range(5):
  num = float(input("Enter a number: "))
  if num > 10:
    count +=1 # Increase the count by 1 every time we get
                      # a number larger than 10
print(count)
```

```python
# To count two things we use two count variables
# Example: Create a program that asks the user to enter 5 numbers and
# count how many of them are greater than 5 and how many of them
# are zero
```

```python
count1 = 0 # count1 starts at 0
count2 = 0 # count2 starts at 0
for k in range(5):
  num = float(input("Enter a number: "))
  if num >5:
    count1 +=1
  if num ==0:
    count2 +=1
print(count1)
print(count2)
```

```python
# Write a program that counts how many of the squares from
# 1^2 to 100^2 end in a 6
count = 0
for k in range(1,101):
  if (k**2)%10 == 6:
    count +=1
print(count)
```

```python
# Summing
# Closely related to counting is summing, where we want to add up
# a bunch of numbers
# Write a program that adds from 1 to 100.
tot = 0
for k in range(1,101):
  tot +=k
print(tot)
```

```python
# Compute the average of 3 numbers:
tot = 0
for k in range(3):
  num = float(input("Enter a number: "))
  tot +=num
print(tot/3)
```

```python
# Swapping
# Incorrect way
x = 2
y = 3
x = y
y = x
print(x)
print(y)
```

```python
# Usual way to swap variables
x = 2
y = 3
hold = x
x = y
y = hold
print(x)
print(y)
```

```python
# Python shortcut
x = 2
y = 3
x,y = y, x
print(x)
print(y)
```

```python
# Write a program that checks if num is prime
num = int(input("Enter a number greater than 1: "))

flag = 0
for k in range(2, num):
  if num%k ==0:
    flag =1
if flag ==0:
  print(num,"is prime")
else:
  print(num,"is not prime")
```

```python
# Some helpful examples
# Write a program that generates 10000 random numbers between
# 1 and 100 and count how many of them are multiple of twelve
from random import randint
count = 0
for k in range(10000):
  num = randint(1,100)
  if num%12 ==0:
    count +=1
print(count)
```

```python
import math
count = 0
for num in range(1, 1001):
  # Check if the square root of the number is an integer
  if math.isqrt(num) ** 2 != num:
    count += 1
print(count)
# math.isqrt(num) returns the largest integer x such that x * x <= num.
# math.isqrt(num) returns the integer part of the square root of num.
```
