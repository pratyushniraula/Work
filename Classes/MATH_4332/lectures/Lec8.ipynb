```python
# Escape characters
# To insert characters that are illegal in a string,
# use an escape character.
# An escape character is a backslash \ followed by
# the character you want to insert.
# An example of an illegal character is a single
# quote inside a string that is surrounded by single quotes:
# E.g
# You will get an error in this
#s= 'A ' inside a string.'
# To fix this problem, use the escape character \':
s= 'A \' inside a string'
print(s)

# same idea applies to a double quote inside
# a string that is surrounded by double quotes:

#s= "A " inside a string."
# To fix this, use \"
s= "A \" inside a string."
print(s)
```

```python
# Example 7
# A simple and very old method of sending secret messages is the
# substitution cipher(coded message). Basically, each letter of the alphabet
# gets replaced by another letter of the alphabet,
# say every a gets replaced with a z,
# and every b gets replaced with a y, etc.
# Write a program to implement this.
alphabet = ascii_lowercase
key = alphabet[::-1]
secret_message = input('Enter your message: ')
secret_message = secret_message.lower()
for c in secret_message:
   if c.isalpha():
    print(key[alphabet.index(c)],end='')
   else:
      print(c, end='')
# Enter:
# Students, exam 1 is on Thursday, October 2
```

```python
#4
# (a) Mimic in operator
s=input("Enter a string: ")
letter = input("Enter a letter: ")
if letter.isalpha():
  flag=0
  for k in range(len(s)):
    if s[k]==letter:
      flag=1
  if flag==1:
    print("The string contains the letter",letter)
  else:
    print("The string does not contain the letter",letter)
else:
  print(letter,"is not a letter.")
```

```python
# (b) Mimic count operator
s=input("Enter a string: ")
letter = input("Enter a letter: ")
tot=0
if letter.isalpha():
  for k in range(len(s)):
    if s[k]==letter:
      tot+=1
  print(tot)
else:
  print(letter,"is not a letter.")
#Check
print(s.count(letter)==tot)
```

```python
# (c) Mimic the index method
s=input("Enter a string: ")
letter = input("Enter a letter: ")

if letter.isalpha():
  count=0
  flag=0
  for k in range(len(s)):
    if s[k]==letter:
      flag = 1
      count +=1
      if count == 1:
        index_first_oc=k
  if flag==1:
    print("The index of the first occurrence of",letter,"is",index_first_oc)
  else:
    print("String does not contain",letter)
else:
  print(letter,"is not a letter.")
# Check
if letter in s:
  print(s.index(letter)==index_first_oc)
```
