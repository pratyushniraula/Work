```python
# Mutability
# Given
L = [3,8,99,123]
# We can change its first element by
L[0]=7
print(L)
# This is possible because lists are mutable
# ie their contents can change. Same for dictionaries
```

```python
# Given
s = "Hello"
# We cannot change the first character of s by s[0]="D" because
# strings are immutable. They cannot be changed
# However we can create a new string based on s
s = "D" + s[1:]
print(s)
```

```python
# Making copies
# Here we make a copy of s and then change s
s = "Hello"
cop = s
s = s + "&&&"
print("s is now:",s," Copy is:",cop)
```

```python
# Here the copy is modified when L is modified
L = [3,45,78]
cop = L
L[0]= 11
print("L is now: ",L," Copy is: ", cop)
```

```python
# Here the copy is not modified when L is modified
L = [3,45,78]
copy = L
L = L + [3]
print("L is now: ",L," Copy is: ", copy)
```

```python
# The key to understanding these examples is references
# References
s = "Hello" # a string object Hello is created and s points to that object
copy = s # both copy and s point to the same string object Hello
# the string object Hello now has two references (copy and s)
s = s + "&&&" # Here a new string object "Hello&&&" is created
# since it has been assigned to s, s is now its reference
print(copy)
print(s)
```

```python
L = [3,45,78] # list object is created with L as reference
copy = L # both copy and L point to the same list
L[0] = 11 # L is changed in place to [11,45,78]
# No new list object is created
print(copy)
print(L)
```

```python
L = [3,45,78] # list object is created with L as reference
copy = L # both copy and L point to the same list
L += [3] # same as L = L + [3]
L = L + [3] # Here a new list object [3,45,78,3] is created
# since it has been assigned to L, L is now its reference
print(copy)
print(L)
```

```python
L = [3,45,78]
copy = L[:] # new list object is created and copy is its reference
L[0]=11 # L is changed to [11,45,78]
print(copy)
print(L)
```

```python
L = [3,45,78]
copy = L.copy() # new list object is created and copy is its reference
L[0]=11 # L is changed to [11,45,78]
print(copy)
print(L)
```

```python
# Tuples
# A tuple is essentially an immutable list
#Example
L = [1,2,3] # list
t = (1,2,3) # tuple
# indexing and slicing work the same way as with list
# t[i]: element located at index i
print(t[1])
# t[:k]: tuple of first k elements (k>0)
print(t[:2])
# t[k:]: tuple of last k elements (k<0)
print(t[-2:])
# length
print(len(t))
# Tuples have count and index methods
t1=(2,4,5,2,5,2)
print(t1.count(2))
```

```python
# Converting objects to tuples
# converting list to tuples
t1 = tuple([3,7,99])
print(t1)
```

```python
# converting string to tuple
t2 = tuple("Hello")
print(t2)
```

```python
# Note:
# Tuple with one element is defined like this: (3,)
print((3,))
# (3) will not work because it just evaluates to the number 3
# Note that (3) != (3,)
print((3) != (3,))
print((3) == (3,))
print(len((3,)))
```

```python
# Empty tuple
t = ()
print(t)

t += (5,)
print(t)
t+= (10,20)
print(t)
# Tuples are immutable
# t[0] = 10 will give error
```

```python
# sets
# sets work like mathematical sets
# They are a lot like list with repeats
# Examples of sets:
S = {3,4,5}
print(S)
```

```python
# Empty set
S = set()
print(S)
# Recall: the notation S={} is used for an empty dictionary
# Note: set()!={}
print(set()=={})
print(set()!={})
# converting list to set
S1 = set([4,3,5,4,5,4])
print(S1)
```

```python
# string to set
S2 = set("Hello there!")
print(S2)
# Indexing has no meaning for sets
```

```python
# Working with sets
# Union: |
S1 = {3,4,5}
S2 = {5,7,9}
S3 = S1 | S2 # S3 = {3,4,5,7,9}
print(S3)
```

```python
# Intersection : &
S1 = {3,4,5}
S2 = {5,7,9}
S3 = S1 & S2 # S3 = {5}
print(S3)
```

```python
# Difference: -
S1 = {3,4,5}
S2 = {5,7,9}
S3 = S1 - S2 # All elements in S1 that are not in S2: {3,4}
print(S3)
S4 = S2 - S1 # All elements in S2 that are not in S1: {7,9}
print(S4)
```

```python
# Symmetric Difference: ^
S1 = {3,4,5}
S2 = {5,7,9}
S3 = S1^S2
# (S1 - S2) | (S2 - S1) = {3,4,7,9}
S4 = (S1 - S2) | (S2 - S1)
print(S4)
# Or S4 = (S1 | S2) - (S1 & S2)
S5 = (S1 | S2) - (S1 & S2)
print(S5)
print(S3)
```

```python
# in operator
print(2 in S1)
print( 5 in S2)
```

```python
# Set methods
# add(x): it adds x to the set
S = {3, 4, 7, 9}
S.add(78)
print(S)
```

```python
# remove(x): removes x from the set
S = {3, 4, 7, 9}
S.remove(3)
print(S)
```

```python
# S.issubset(A): Returns True if S is inside A
A = {5,35,99,4}
S = {9, 4, 7}
S1 = {5,35}
print(S.issubset(A))
print(S1.issubset(A))
print(A.issubset(A))
```

```python
# S.issuperset(A): Returns True if A is inside S
A = {5,35,99,4}
S = {9, 4, 7}
S1 = {5,35}
print(S.issuperset(A))
print(S1.issuperset(A))
print(A.issuperset(A))
```

```python
# sorted
# an iterable: object that you can loop through its contents
# Examples includes lists, tuples, sets, strings and dictionaries
# The sorted function is a built-in function that can be used
# to sort an iterable
# sorted used on iterable produces a sorted list
L = [11,45,23]
M = sorted(L) # creates a new list with M as reference
L[1]="string"
print(L)
print(M)
```

```python
# continue statement:
# we use break to break out of loops
# we use continue to ignore all the code in the loop beyond
# the continue statement and jumps back to the start of the loop,
# advancing the loop counter as necessary
# Example:
L = [1,3,4,3,4,5,3]
l = []
for item in L:
  if item not in l:
    l.append(item)
print(l)
```

```python
# Let us accomplish the same with continue statement
L = [1,3,4,3,4,5,3]
l1 = []
for item in L:
  if item in l1:
    continue
  l1.append(item)
print(l1)
```

```python
# enumerate function
# the built in enumerate function takes an iterable and returns
# a new iterable consisting of the pairs (i,x) where i is an index
# and x is the corresponding element from the iterable
# example
s = "xyzabc"
for index,c in enumerate(s):
  print(index,c)
```

```python
# To obtain list of pairs, use list(enumerate(s))
print(list(enumerate(s)))
```

```python
# Zip function
# The zip function takes two iterables and put them into
# a single iterable that contains the pairs (x,y), where
# x is from the first iterable and y from the second iterable
s = "abc"
L = [1,6,9]
z = zip(s,L)
print(list(z))
```

```python
# Use zip to create dictionary
L1 = ["A","B","C"]
L2 = [6,7,9]
d = dict(list(zip(L1,L2)))
print(d)
```
