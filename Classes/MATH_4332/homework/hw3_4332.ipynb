```python
#1
import random

# List of questions and answers
questions = [
    "What is 2 + 2?",
    "What is the capital of France?",
    "What color is the sky?",
    "What is 5 * 6?",
    "What programming language are we learning?"
]

answers = [
    "4",
    "paris",
    "blue",
    "30",
    "python"
]

# Pick 3 random question indices
indices = random.sample(range(len(questions)), 3)

score = 0
for i in indices:
    user_answer = input(questions[i] + " ").strip().lower()
    if user_answer == answers[i]:
        print("Correct!")
        score += 1
    else:
        print("Wrong!")
print(f"You got {score} out of 3 correct.")
```

```python
#2
s = input("Enter a phone number: ").strip()

parts = s.split("-")
valid = False

if len(parts) == 4 and parts[0] == "1":
    # 1-abc-def-hijk
    a, b, c = parts[1], parts[2], parts[3]
    if a.isdigit() and b.isdigit() and c.isdigit() and len(a) == 3 and len(b) == 3 and len(c) == 4:
        valid = True
elif len(parts) == 3:
    # abc-def-hijk
    a, b, c = parts[0], parts[1], parts[2]
    if a.isdigit() and b.isdigit() and c.isdigit() and len(a) == 3 and len(b) == 3 and len(c) == 4:
        valid = True

print("Valid" if valid else "Invalid")
```

```python
#4b
import random
import string

raw = input("Enter a sentence: ").strip()

# ensure we have words and strip one trailing period if present
has_period = raw.endswith(".")
core = raw[:-1].strip() if has_period else raw
words = core.split()

# remember original first word (case-insensitive compare)
orig_first = words[0]
orig_first_lower = orig_first.lower()

random.shuffle(words)

# lowercase everything first for consistency
words = [w.lower() for w in words]

# capitalize the new first word
if words:
    words[0] = words[0].capitalize()

# if the original first word appears later, keep it lowercase (not capitalized mid-sentence)
for i in range(1, len(words)):
    if words[i].lower().strip(string.punctuation) == orig_first_lower:
        words[i] = words[i].lower()

# add a period at the end
out = " ".join(words).strip()
if not out.endswith("."):
    out += "."

print(out)
```
