```python
#leibniz equation for pi = 4 * (1 - 1/3 + 1/5 - 1/7 + 1/9 - ...)
#Use numpy arrays to find an approximation of π using the first 3,333 terms of Leibniz’s infinite sum.
#Display the approximated value.
#Up to how many decimal places is the approximation accurate?
import numpy as np
n_terms = 3333
denominators = np.arange(1, 2*n_terms, 2)
signs = np.array([1 if i % 2 == 0 else -1 for i in range(n_terms)])
leibniz_sum = np.sum(signs / denominators)
pi_approximation = 4 * leibniz_sum
print(f"Approximation of π using the first {n_terms} terms of Leibniz's series: {pi_approximation}")
#how many decimal places is the approximation accurate?
pi_actual = np.pi
accuracy = abs(pi_actual - pi_approximation)
decimal_places = 0
while accuracy < 0.5 * 10**(-decimal_places):
    decimal_places += 1
print(f"The approximation is accurate up to {decimal_places} decimal places.")
```

```python
#Wallis’s equation for pi = 2 * (2/1 * 2/3 * 4/3 * 4/5 * 6/5 * 6/7 * ...)
#Use numpy arrays to find an approximation of π using 55,555 products of Wallis’ infinite product.
#Display the approximated value.
#Up to how many decimal places is the approximation accurate?
import numpy as np

n_products1 = 55555
n = np.arange(1, n_products1 + 1, dtype=np.float64)
ratios = (4*n*n) / (4*n*n - 1)            # Wallis terms
pi_approximation_wallis = 2 * np.prod(ratios)

print(f"Approx π with {n_products1} Wallis products: {pi_approximation_wallis:.18f}")

pi_actual = np.pi
accuracy_wallis = abs(pi_actual - pi_approximation_wallis)

decimal_places_wallis = 0
while accuracy_wallis < 0.5 * 10**(-decimal_places_wallis):
    decimal_places_wallis += 1

print(f"Accurate to {decimal_places_wallis} decimal places.")
```

```python
# Write a function my_fun_plotter(f, x) where f is a lambda function and x is an array. The function
# should plot f evaluated at x. Remember to label the x- and y-axis.
import matplotlib.pyplot as plt

def my_fun_plotter(f, x):
    y = f(x)
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title("Plot of f(x)")
    plt.grid()
    plt.show()
#Test the function with f(x) = x^2 and x values from -10 to 10.
my_fun_plotter(lambda x: np.sqrt(x) + np.exp(np.sin(x)), np.linspace(0, 2*np.pi, 100))
```

```python
#Write a function my_poly_plotter(n,x) that plots the polynomials pk(x) = x^k
#for k = 1, 2, · · · n. Make
#sure you plot the axis labels and a title.
import matplotlib.pyplot as plt

def my_poly_plotter(n, x):
    for k in range(1, n + 1):
        plt.plot(x, x**k, label=f"p{k}(x) = x^{k}")
    plt.xlabel("x")
    plt.ylabel("p_k(x)")
    plt.title("Plot of Polynomials p_k(x) = x^k")
    plt.legend()
    plt.grid()
    plt.show()

my_poly_plotter(5, np.linspace(-1, 1, 200))
```

```python
#Write a function my_surface_plotter(f, x, y, option) where f is a function object f(x,y). The function my_surface_plotter should produce a 3D surface plot of f(x,y) using plot_surface if option is the
#string ’surface’. It should produce a contour plot of f(x,y) if the option is the string ’contour’. Assume
#that x and y are one-dimensional arrays or lists. Remember to give the plot a title and axis labels.
from mpl_toolkits.mplot3d import Axes3D
def my_surface_plotter(f, x, y, option):
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    
    fig = plt.figure()
    
    if option == 'surface':
        ax = fig.add_subplot(111, projection='3d')
        ax.plot_surface(X, Y, Z, cmap='viridis')
        ax.set_title("3D Surface Plot")
        ax.set_xlabel("X-axis")
        ax.set_ylabel("Y-axis")
        ax.set_zlabel("f(X,Y)")
    elif option == 'contour':
        ax = fig.add_subplot(111)
        contour = ax.contourf(X, Y, Z, cmap='viridis')
        fig.colorbar(contour)
        ax.set_title("Contour Plot")
        ax.set_xlabel("X-axis")
        ax.set_ylabel("Y-axis")
    else:
        raise ValueError("Option must be either 'surface' or 'contour'")
    
    plt.show()

f = lambda x, y: np.cos(y)*np.sin(np.exp(x))
my_surface_plotter(f, np.linspace(-1, 1, 20), np.linspace(-2, 2, 40), 'surface')
my_surface_plotter(f, np.linspace(-1, 1, 20), np.linspace(-2, 2, 40), 'contour')
```

```python
#Let −4 ≤ x ≤ 4 and −3 ≤ y ≤ 3 and z(x, y) =xy((x^2-y^2)/x^2+y^2)
#Create arrays x and y with 100 evenly spaced
# points over the interval. Create meshgrids X and Y for x and y using the meshgrid function. Compute
# the matrix Z from X and Y . Create a 1×2 subplot where the first figure is the 3D surface Z plotted using
# plt.plot_surface and the second figure is the 3D wireframe plot using plt.plot_wireframe, respectively.
# Give each axis a title and axis labels. Save your output as a pdf.
x = np.linspace(-4, 4, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = X * Y * ((X**2 - Y**2) / (X**2 + Y**2))
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title("3D Surface Plot of z(x,y)")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.set_zlabel("z(X,Y)")
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(X, Y, Z, color='r')
ax2.set_title("3D Wireframe Plot of z(x,y)")
ax2.set_xlabel("X-axis")
ax2.set_ylabel("Y-axis")
ax2.set_zlabel("z(X,Y)")
plt.tight_layout()
plt.savefig("surface_wireframe_plots.pdf")
plt.show()
```
