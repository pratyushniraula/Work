```python
#1. Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds. [Hint: Use the // operator to get minutes and the % operator to get seconds.]
x = int(input("Enter number of seconds: "))
minutes = x // 60
seconds = x % 60
print(x, " seconds is ", minutes, " minutes and ", seconds, " seconds.", sep='')
```

```python
#2.a Write a program that asks the user for a number and prints out the factorial of that number. [Hint: Use the math module.]
import math
x = int(input("Enter a number: "))
print("The factorial of ", x, " is ", math.factorial(x), ".", sep='', end='\n\n---------------\n\n')

#2.b Write a program that asks the user to enter an angle in degrees and prints out the sine of that angle. [Hint: Use the math module.]
x = float(input("Enter an angle in degrees: "))
radians = math.radians(x)
print("The sine of ", x, " degrees is ", math.sin(radians), ".", sep='')
```

```python
#3. Ask the user for a temperature. Then ask them what units, Celsius (use 0 for Celsius) or Fahrenheit (use 1 for Fahrenheit ), the temperature is in. Your program should convert the temperature to the other unit.
unit = int(input("Enter 0 for Celsius or 1 for Fahrenheit: "))
temp = float(input("Enter a temperature: "))
if unit == 0:
    converted = (temp * 9/5) + 32
    print(temp, " degrees Celsius is ", converted, " degrees Fahrenheit.", sep='')
elif unit == 1:
    converted = (temp - 32) * 5/9
    print(temp, " degrees Fahrenheit is ", converted, " degrees Celsius.", sep='')
else:
    print("Invalid unit entered. Please enter 0 for Celsius or 1 for Fahrenheit.")
```

```python
#4. Write a program that asks the user how many credits they have taken. If they have taken 23 or less, print that the student is a freshman. If they have taken between 24 and 53, print that they are a sophomore. The range for juniors is 54 to 83, and for seniors it is 84 and over.
credits = int(input("Enter number of credits taken: "))
if credits <= 23:
    print("The student is a freshman.")
elif 24 <= credits <= 53:
    print("The student is a sophomore.")
elif 54 <= credits <= 83:
    print("The student is a junior.")
elif credits >= 84:
    print("The student is a senior.")
else:
    print("Invalid number of credits entered.")
```

```python
#5. A year is a leap year if it is divisible by 4, except that years divisible by 100 are not leap years unless they are also divisible by 400. Write a program that asks the user for a year and prints out whether it is a leap year or not.
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, " is a leap year.", sep='')
else:
    print(year, " is not a leap year.", sep='') 
```
