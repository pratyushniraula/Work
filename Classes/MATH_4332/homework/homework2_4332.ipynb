```python
#1 Write a program that asks the user to enter a word and then capitalizes every other letter of that word. 
# So if the user enters rhinoceros, the program should print rHiNoCeRoS.

word = input("Enter a word: ")
new_word = ""
for i in range(len(word)):
    if i % 2 == 1:
        new_word += word[i].upper()
    else:
        new_word += word[i]

print(word, " has been made to: ", new_word)
```

```python
#2 Write a program that asks the user to enter their name in lowercase and then capitalizes the first letter of each word of their name.
name = input("Enter your name in lowercase: ")
new_name = ""
capitalize_next = True
for char in name:
    if char == ' ':
        new_name += char
        capitalize_next = True
    elif capitalize_next:
        new_name += char.upper()
        capitalize_next = False
    else:
        new_name += char

print(name, " has been capitalized to: ", new_name)
```

```python
#Ask the user to enter 10 test scores. Write a program to do the following:

#1. Print out the smallest and the second smallest scores
#2 If any of the scores is greater than 100, then after all the scores have been entered, print a message warning the user that a value over 100 has been entered.
#3 Drop the two lowest scores and print out the average of the rest of them

scores = []
for i in range(10):
    score = int(input("Enter test score " + str(i+1) + ": "))
    scores.append(score)
    if score > 100:
        print("Warning: A score greater than 100 has been entered.")

print("The two smallest scores are: ", end='')
for i in range(2):
    smallestScoreIndex = 0
    for i in range(len(scores)):
        if scores[i] < scores[smallestScoreIndex] and scores[i] != -1:
            smallestScoreIndex = i
    print(scores[smallestScoreIndex], end=' ')
    scores[smallestScoreIndex] = -1
new_Score = 0
for score in scores:
    if score != -1:
        new_Score += score
average = new_Score / 8
print("\nThe average after dropping the two lowest scores is: ", average)
            
```

```python
#Write a program that asks the user to guess a random number between 1 and 10. If they guess right, they
#get 10 points added to their score, and they lose 1 point for an incorrect guess. Give the user five numbers
#to guess and print their score after all the guessing is done.
import random
score = 0
for i in range(5):
    randomNumber = random.randint(1, 10)
    guess = int(input("Guess a number between 1 and 10: "))
    if guess == randomNumber:
        score += 10
        print("Correct! You get 10 points.")
    else:
        score -= 1
        print("Wrong! The correct number was ", randomNumber, ". You lose 1 point.", sep='')
print("Your final score is: ", score)
```

```python
# A simple way of encrypting a message is to rearrange its characters. One way to rearrange the characters
# is to pick out the characters at even indices, put them first in the encrypted string, and follow them by
# the odd characters. For example, the string message would be encrypted as msaeesg because the even
# characters are m, s, a, e (at indices 0, 2, 4, and 6) and the odd characters are e, s, g (at indices 1, 3, and
# 5).
# (a) Write a program that asks the user for a string and uses this method to encrypt the string.
# (b) Write a program that decrypts a string that was encrypted with this method.
message = input("Enter a message to encrypt: ")
even_chars = ""
odd_chars = ""
for i in range(len(message)):
    if i % 2 == 0:
        even_chars += message[i]
    else:
        odd_chars += message[i]
encrypted_message = even_chars + odd_chars
print("The encrypted message for", message, " is: ", encrypted_message)
```

```python
# (b) Write a program that decrypts a string that was encrypted with this method.
encrypted_message = input("Enter a message to decrypt: ")
mid_index = len(encrypted_message) // 2 + len(encrypted_message) % 2
even_chars = encrypted_message[:mid_index]
odd_chars = encrypted_message[mid_index:]
decrypted_message = ""
for i in range(mid_index):
    decrypted_message += even_chars[i]
    if i < len(odd_chars):
        decrypted_message += odd_chars[i]
print("The decrypted message is: ", decrypted_message)
```
