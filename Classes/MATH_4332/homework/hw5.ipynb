```python
# Write a function called closest() that takes a list of integers and another integer n and returns the integer
# in the list that is the closest to n. If two integers are tied for the closest, then return the smaller of the
# two.
def closest(int_list, n):
    if not int_list:
        return None  # Return None if the list is empty

    closest_num = int_list[0]
    smallest_diff = abs(closest_num - n)

    for num in int_list:
        current_diff = abs(num - n)
        if current_diff < smallest_diff:
            smallest_diff = current_diff
            closest_num = num
        elif current_diff == smallest_diff and num < closest_num:
            closest_num = num

    return closest_num

closest([4, 1, 7, 10], 6)  # Should return 7
```

```python
# Write a function called names() that returns a list of all of the names in the following string:
# s = """Anna is 7 years old, and her sister Olivia is 2 years old.
# Evelyn and Paul, their parents, have 3 kids."""
# Also find len(names()).

def names():
    s = """Anna is 7 years old, and her sister Olivia is 2 years old.
    Evelyn and Paul, their parents, have 3 kids."""
    name_list = []
    for line in s.splitlines():
        for word in line.split():
            if word[0].isupper() and word[1:].islower():
                name_list.append(word)
    return name_list

print(names())
print(len(names()))
```

```python
# The file grades.txt contains a line separated list of people with their grade in a class. Write a function
# called grades() that returns a list of just those students who received a B in the course. Also find
# len(grades()).

def grades():
    with open("grades.txt", "r") as f:
        lines = f.readlines()
    b_students = []
    for line in lines:
        name, grade = line.split()
        if grade == "B":
            b_students.append(name)
    return b_students

print(grades())
print(len(grades()))
```

```python
# Consider the file logdata.txt. This file records the access a user makes when visiting a web page. Each
# line of the logdata.txt has the following items:
# • a host (e.g., ’146.204.224.152’)
# • a user_name (e.g., 'feest6811' note: sometimes the username is missing! In this case, use '-' as
# the value for the username.)
# • the time a request was made (e.g., '21/Jun/2019:15:45:24 -0700')
# • the post request type (e.g., 'POST /incentivize HTTP/1.1' note: not everything is a POST!).
                         
# Write a function called logs() that reads through the lines of the file and converts them into a list of
# dictionaries, where each dictionary looks like the following:
# example_dict = {"host":"146.204.224.152",
# "user_name":"feest6811",
# "time":"21/Jun/2019:15:45:24 -0700",
# "request":"POST /incentivize HTTP/1.1"}
# 
# Also find len(logs()).
def logs():
    log_list = []
    with open("logdata.txt", "r") as f:
        lines = f.readlines()
    for line in lines:
        parts = line.split()
        host = parts[0]
        user_name = parts[1]
        time = parts[3] + " " + parts[4]
        request = " ".join(parts[5:8])
        log_entry = {
            "host": host,
            "user_name": user_name,
            "time": time.strip("[]"),
            "request": request.strip('"')
        }
        log_list.append(log_entry)
    return log_list
```

```python
# # Recall that if s is a string, then s.find('a') will find the location of the first a in s. The problem is that
# it does not find the location of every a. Write a function called findall that given a string and a single
# character, returns a list containing all of the locations of that character in the string. It should return an
# empty list if there are no occurrences of the character in the string.

def findall(s, char):
    locations = []
    for index, c in enumerate(s):
        if c == char:
            locations.append(index)
    return locations

print(findall("banana", "a"))  # Should return [1, 3, 5]
print(findall("apple", "z"))   # Should return []
print(findall("", "a"))        # Should return []
```

```python
# The digital root of a number n is obtained as follows: Add up the digits of n to get a new number.
# Add up the digits of that to get another new number. Keep doing this until you get a number that has
# only one digit. That number is the digital root. For example, if n = 45893, we add up the digits to
# get 4 + 5 + 8 + 9 + 3 = 29. We then add up the digits of 29 to get 2 + 9 = 11. We then add up the
# digits of 11 to get 1 + 1 = 2. Since 2 has only one digit, 2 is our digital root. Write a function that
# returns the digital root of an integer n. [Note: there is a shortcut, where the digital root is equal to
# n mod 9, but do not use that here.]

def digital_root(n):
    while n >= 10:
        n = sum(int(digit) for digit in str(n))
    return n

print(digital_root(45893))  # Should return 2
print(digital_root(123456))  # Should return 3
print(digital_root(9))       # Should return 9
```

```python
# #Write a function called closest that takes a list of numbers L and a number n and returns the largest
# element in L that is not larger than n. For instance, if L = [1,6,3,9,11] and n = 8, then the function
# should return 6, because 6 is the closest thing in L to 8 that is not larger than 8. Don’t worry about if all
# of the things in L are smaller than n.

def closest(L, n):
    closest_num = None
    for num in L:
        if num <= n:
            if closest_num is None or num > closest_num:
                closest_num = num
    return closest_num

print(closest([1, 6, 3, 9, 11], 8))  # Should return 6
print(closest([10, 20, 30], 25))      # Should return 20
print(closest([5, 15, 25], 5))        # Should return 5
```
